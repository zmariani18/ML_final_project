```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import os 
import matplotlib.pyplot as plt

def clean_FIPS(x):
    if len(str(x)) != 5: 
        return '0' + str(x)
    else: 
        return str(x)
```

```python
df = pd.read_csv('/Users/zachmariani/ECMA/ECMA_Final_Project/Data/X_C_y_clean.csv')
df['FIPS'] = df['FIPS'].apply(clean_FIPS)
df['date'] = pd.to_datetime(df['date'])
df.set_index(['FIPS', 'date'], inplace=True)
df['C_state_code'] = df.index.get_level_values(0).str[0:2]
```
